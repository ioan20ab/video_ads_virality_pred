```python
from datetime import timedelta
import cv2
import numpy as np
import os
```

```python
SAVING_FRAMES_PER_SECOND = 10
```

```python
def format_timedelta(td):
    """Utility function to format timedelta objects in a cool way (e.g 00:00:20.05) 
    omitting microseconds and retaining milliseconds"""
    result = str(td)
    try:
        result, ms = result.split(".")
    except ValueError:
        return (result + ".00").replace(":", "-")
    ms = int(ms)
    ms = round(ms / 1e4)
    return f"{result}.{ms:02}".replace(":", "-")


def get_saving_frames_durations(cap, saving_fps):
    """A function that returns the list of durations where to save the frames"""
    s = []
    # get the clip duration by dividing number of frames by the number of frames per second
    clip_duration = cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS)
    # use np.arange() to make floating-point steps
    for i in np.arange(0, clip_duration, 1 / saving_fps):
        s.append(i)
    return s


def substring_after(s, delim):
    return s.partition(delim)[2]

def substring_before(s, delim):
    return s.partition(delim)[0]
```

```python
def make(video_file, namefile):
#     filename, _ = os.path.splitext(video_file)
#     filename += "-opencv"
#video_file is the path of the video in _DATA 4

    _path = "D:\_CBS\_Thesis\_Project\Video ads\_DATA 5"
    target_path = os.path.join(_path, namefile)
    # make a folder by the name of the video file
    if not os.path.isdir(target_path):
        os.mkdir(target_path)
        
    #CHANGE PATH
    
    
#     delim1 = "_DATA 4"
#     video_file1 = substring_after(video_file,delim)
#     print("delimited video file: "+video_file)
#     video_file2= substring_after(video_file1,"\")
    path_to_file = "D:/_CBS/_Thesis/_Project/Video ads/_DATA 4"+"/"+namefile
    os.chdir(path_to_file)
    print(os.getcwd())
    video_load = namefile+".mp4"
    
    
    
    
    # read the video file    
    cap = cv2.VideoCapture(video_load)
    
    if not cap.isOpened():
        print("Cannot open file")
    # get the FPS of the video
    fps = cap.get(cv2.CAP_PROP_FPS)
    # if the SAVING_FRAMES_PER_SECOND is above video FPS, then set it to FPS (as maximum)
    saving_frames_per_second = min(fps, SAVING_FRAMES_PER_SECOND)
    # get the list of duration spots to save
    saving_frames_durations = get_saving_frames_durations(cap, saving_frames_per_second)
    # start the loop
    count = 0
    while True:
#         print("In while")
        is_read, frame = cap.read()
        if not is_read:
#             print("if not is_read")
            # break out of the loop if there are no frames to read
            break
        # get the duration by dividing the frame count by the FPS
        frame_duration = count / fps
        try:
#             print("first try")
            # get the earliest duration to save
            closest_duration = saving_frames_durations[0]
        except IndexError:
#             print("except IndexError")
            # the list is empty, all duration frames were saved
            break
        if frame_duration >= closest_duration:
            # if closest duration is less than or equals the frame duration, 
            # then save the frame
#             print("In second if")
            frame_duration_formatted = format_timedelta(timedelta(seconds=frame_duration))
#             cv2.imwrite(os.path.join(namefile+"~", f"frame{frame_duration_formatted}.jpg"), frame) 
            
            cv2.imwrite(os.path.join(target_path,namefile+"~"+f"frame{frame_duration_formatted}.jpg"), frame)
            
            # drop the duration spot from the list, since this duration spot is already saved
            try:
#                 print("In second try")
                saving_frames_durations.pop(0)
            except IndexError:
#                 print("In second except IndexError")
                pass
        # increment the frame count
        count += 1
```
