```python
import pandas as pd

from deepface import DeepFace

import matplotlib.pyplot as plt

import os
```

```python
columns1= ["url", "url_path", "frame", "frame_path"]

backends = [ "opencv", "ssd", "dlib", "mtcnn", "retinaface", "mediapipe"]

def substring_after(s,delim):
    return s.partition(delim)[2]

def substring_before(s,delim):
    return s.partition(delim)[0]

def detect_fa(path, backend):
    face = DeepFace.extract_faces(path, detector_backend=backend)
    return face

def facial_analysis(path, backend):
    demographies = DeepFace.analyze(img_path = path, 
        detector_backend = backend)
    return demographies
```

```python
## Z FOLDER
pthz = "E:\_CBS\_Thesis\_Project\Video ads\_DATA 5\Z"

Z_urls= [ f for f in os.listdir(pthz) if os.path.isdir(os.path.join(pthz, f)) ]

print(len(Z_urls))
print(Z_urls)
```

```python
columns1 = ["url", "url_frame", "url_path", "frame_path", "emotion", 'dominant_emotion', 'age', 'gender', 'dominant_gender']
Z_df = pd.DataFrame(columns = columns1)

for i in range(len(Z_urls)):
    print("START i url: "+ str(i))
    url = Z_urls[i]
    url_path = os.path.join(pthz, Z_urls[i]) # eg. "E:\_CBS\_Thesis\_Project\Video ads\_DATA 5\Z\Z_hwMtG2xCs"
    
    url_frames = [ f for f in os.listdir(url_path) if os.path.isfile(os.path.join(url_path, f)) ]
    
    # Now we have every frame of the specific url
    # Loop through the frames of the url
    for x in range(len(url_frames)):
        url_frame = url_frames[x]
        # make the frame path
        frame_path = os.path.join(url_path, url_frames[x])
        
        # now TRY EXCEPT facial analysis
        
        try:
            result = facial_analysis(frame_path,backends[4] )
            #check if need to loop through faces or not
            
            if len(result)>1:
                
                #loop through faces to save into data frame
                for y in range(len(result)):
                    emotion =          result[y]["emotion"]
                    dominant_emotion = result[y]['dominant_emotion']
                    age =              result[y]['age']
                    gender =           result[y]['gender']
                    dominant_gender =  result[y]['dominant_gender']
                    
                    
                    
                    new_row = {'url': str(url), 'url_frame': str(url_frame), 'url_path': str(url_path),
                               'frame_path': frame_path,"emotion" : emotion, 'dominant_emotion' :  dominant_emotion,
                              'age' : age, 'gender' : gender, 'dominant_gender' : dominant_gender}
                    Z_df.loc[len(Z_df)] = new_row
                    print("saved 1 face in loop")
                  
                    
                    
                
            else:
                
                
                # save 1 face in dataframe
                emotion =          result[y]["emotion"]
                dominant_emotion = result[y]['dominant_emotion']
                age =              result[y]['age']
                gender =           result[y]['gender']
                dominant_gender =  result[y]['dominant_gender']
                    
                    
                    
                new_row = {'url': str(url), 'url_frame': str(url_frame), 'url_path': str(url_path),
                            'frame_path': frame_path,"emotion" : emotion, 'dominant_emotion' :  dominant_emotion,
                            'age' : age, 'gender' : gender, 'dominant_gender' : dominant_gender}
                
                Z_df.loc[len(Z_df)] = new_row
                print("saved 1 face")
                
                
            
            
        except:
            continue


path3 = "E:\_CBS\_Thesis\_Project\Video ads\_DATA 5\Z_df.txt"
Z_df.to_csv(path3, header=None, index=None, sep=';', mode='w')
```
