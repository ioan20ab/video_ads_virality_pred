```python
import os

# Directory 

  
# Parent Directory path 
parent_dir = "D:/_CBS/_Thesis/_Project/Video ads/_DATA/"

for index, row in df2.iterrows():
    video_directory = df2.loc[index, "video id"]
#     print(video_directory)
    path = os.path.join(parent_dir, video_directory) 
    os.mkdir(path)
    
```

```python
from pytube import YouTube

# # YouTube('https://youtu.be/2lAe1cqCOXo').streams.first().download()
# yt = YouTube('https://www.youtube.com/watch?v=1KkVdbUr250&ab_channel=MalalaFund')

# test

direct_1 = "D:/_CBS/_Thesis/_Project/Video ads/_DATA/"

df_agerestrict = pd.DataFrame(columns = ['video id', 'Link 1', 'Link 2', 'Works', 'Has Face'])

for index, row in df2.iterrows():
    
    vid_id = row["video id"]
    ext = ".mp4"
    file_name = vid_id+ext
    
    path = direct_1 + file_name
    
    try:
        yt = YouTube(row["Link 2"],use_oauth=True,allow_oauth_cache=True)
        yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first().download(output_path=path, filename=file_name)
        print("Downloaded:"+ vid_id)
    
    except :
        df_agerestrict.loc[len(df_agerestrict)]= row
        continue

```
